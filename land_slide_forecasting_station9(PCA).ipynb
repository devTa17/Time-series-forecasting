```python
def correlation(data, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = data.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
def split_sequences(sequences, n_steps):
    X, y = list(), list()
    for i in range(len(sequences)):
# find the end of this pattern
        end_ix = i + n_steps
# check if we are beyond the dataset
        if end_ix > len(sequences):
            break
# gather input and output parts of the pattern
        seq_x, seq_y = sequences[i:end_ix, :-1], sequences[end_ix-1, -1]
        X.append(seq_x) 
        y.append(seq_y)
    return array(X), array(y)
```

#Converting features(PCA) to images(CNN)

```python
from scipy import signal
import cv2
```

```python
from skimage.transform import resize
```

```python
window_size = 128
overlap = 64
```

```python
# Define the spectrogram parameters
fs = 10 # Sampling frequency (Hz)
nperseg = 128 # Number of samples per segment
noverlap = nperseg // 2 # Number of samples to overlap between segments
nfft = nperseg # Number of FFT points
```

```python
spectrogram_images = []
displacement_labels = []

# Iterate over the rows of the dataset
for i in range(len(data_pca)):
    # Get the values of the 'rainfall', 'acceleration', and 'velocity' features
    x = data_pca.iloc[i][['PC1', 'PC2', 'PC3']].values
    
    # Compute the spectrogram using the 'displacement' feature as the target
    disp = np.repeat(data_pca.iloc[i]['total_disp'], nperseg)

    f, t, Sxx = signal.spectrogram(disp, fs=fs, nperseg=nperseg, noverlap=noverlap, nfft=nfft)
    
    # Normalize the spectrogram to values between 0 and 1
  
    
    # Reshape x to have a third dimension
    x = x[..., np.newaxis]
    
    # Concatenate the input features to the normalized spectrogram along the channels axis
    img = np.concatenate([x] * 3 + [Sxx], axis=0)
    
    # Resize the image to 128x128 pixels
    img = cv2.resize(img, (128, 128))
    
    # Add the image to the list of spectrogram images
    spectrogram_images.append(img)
    
    # Get the corresponding displacement label and add it to the list of labels
    displacement_labels.append(data_pca.iloc[i]['total_disp'])
```

```python
X = spectrogram_images # replace with the path to your spectrogram dataset
y = displacement_labels # replace with the path to your displacement labels

# Define helper function to plot spectrogram images
# Define helper function to plot spectrogram images
def plot_spectrograms(images, n_cols=5, fig_size=(20, 8)):
    n_rows = int(np.ceil(len(images) / n_cols))
    fig, axs = plt.subplots(n_rows, n_cols, figsize=fig_size)
    for i, ax in enumerate(axs.flat):
        if i < len(images):
            ax.imshow(images[i], cmap='jet')
            ax.set_title(f'Image {i+1}')
        ax.axis('off')
    plt.tight_layout()

# Select the first 10 images and plot them
images = X[:10]
plot_spectrograms(images)
```
